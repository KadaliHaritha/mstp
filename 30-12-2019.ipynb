```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
df.head()
```

```python
x=df["Age"].values
y=df["Duration"].values
plt.title("data",color='r')
plt.xlabel("Age")
plt.ylabel("Duration")
plt.bar(x,y,color='y')
plt.show()
```

```python
x=df["Credit amount"].values
plt.title("data",color='g')
plt.pie(x)
plt.show()
```

```python
x=df["Age"].values
y=df["Job"].values
z=df["Credit amount"].values
a=df["Duration"].values
c=['g','b','r','k','y']
plt.stackplot(x,y,z,a,colors=c)
plt.show()
```
